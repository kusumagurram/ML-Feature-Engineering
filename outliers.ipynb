```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
#create Data Frame
houses = pd.DataFrame()
houses['Price'] =[532446,352628,555000,463589]
houses['Bathrooms'] = [2,3,4,116]
houses['Square_feet'] = [1500,2500,1700,48000]
houses
```

# Drop

```python
houses[houses['Bathrooms'] < 20]
```

# mark

```python
#create feature based on boolean condition
houses['Outlier'] = np.where(houses['Bathrooms'] < 20,0,1)
houses
```

# Rescale

```python
#Log Feature
houses['Log_Of_Square_feet'] = [np.log(x) for x in houses['Square_feet']]
houses
```
